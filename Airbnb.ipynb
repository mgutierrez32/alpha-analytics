**Detect and Drop Outliers**

```python
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(listing.price))
print(z)
```

```python
threshold = 3
outlier = np.where(z > threshold)
print(outlier)
```

```python
listing = listing.drop(listing.index[outlier])
```

```python
z1 = np.abs(stats.zscore(calendar.price))
print(z1)
```

```python
outlier1 = np.where(z1 > threshold)
print(outlier1)
```

```python
calendar = calendar.drop(calendar.index[outlier1])
```
