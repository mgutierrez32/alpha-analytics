```python
import pandas as pd
import numpy as np
```

```python
listings = pd.read_csv("listing_clean_2Z_roomtype.csv")
```

```python
listings.head()
```

```python
neighborhoods = listings["neighbourhood"]
median_price = listings["price"]

for i, hood in enumerate(neighborhoods):
    median_price[i] = np.median(listings[listings["neighbourhood"]==hood]["price"])
```

```python
listings.insert(9,"neighborhood_median", median_price)
```

```python
listings.head()
```

```python
listings.to_csv("listing_final.csv")
```
